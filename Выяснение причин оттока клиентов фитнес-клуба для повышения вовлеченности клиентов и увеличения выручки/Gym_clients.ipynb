### Описание проекта


Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.

Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Как понять, что клиент больше не с вами? Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо.

Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на Бали и по приезде обязательно продолжит ходить на фитнес. Однако чаще бывает наоборот. Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.

Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Наша задача — провести анализ и подготовить план действий по удержанию клиентов.

## Построить модель прогнозирования оттока клиентов

### Разобьём данные на обучающую и валидационную выборку

```python
X = df.drop('churn', axis = 1)
y = df['churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Обучение модели на нашей выборке

#### Логистическая регрессия

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
probabilities =  model.predict_proba(X_test)[:,1]
```

```python
print('Accuracy | Нет баланса классов: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, predictions)))
```

#### Случайный лес

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)

model_forest = RandomForestClassifier(n_estimators = 100, random_state = 0) 
model_forest.fit(X_train_st, y_train)
predictions_forest = model_forest.predict(X_test_st) 
probabilities_forest = model_forest.predict_proba(X_test_st)[:,1]
```

```python
print('Accuracy | Нет баланса классов: {:.2f}'.format(accuracy_score(y_test, predictions_forest)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions_forest)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions_forest)))
print('F1: {:.2f}'.format(f1_score(y_test, predictions_forest)))
```

**Вывод:**
Модели показали примерно одинаковый результат, но случайный лес всё-таки немного лучше.

## Сделать кластеризацию клиентов

```python
X_st = scaler.fit_transform(X)
linked = linkage(X_st, method='ward')
```

```python
plt.figure(figsize=(15, 10))
dendrogram(linked, orientation='top')
plt.title('Кластеризация клиентов', fontsize=15)
plt.show()
```

Визуально выделяются 4 крупных кластера. Обучим модель алгоритмом K-Means для числа крастеров n = 5.

```python
km = KMeans(n_clusters = 5, random_state = 0)
labels = km.fit_predict(X_st)
df['cluster_km'] = labels
df.head()
```

Сгруппируем данные по кластерам.

```python
df.groupby('cluster_km').mean().T
```

**Кластер 0**

**Самый низкий отток - 2,8%**. Живут рядом с клубом, работают в компании-партнёре, пришли с другом, абонементы **на долгий срок**, посещают групповые чаще клиентов из други кластеров, уже давно являются клиентами клуба.

**Кластер 1**

Отток - **26,8%**. Тоже живут недалеко от клуба, пришли без друзей и не по партнёрской программе, средний срок абонемента, частота тоже средняя.

**Кластер 2**

Отток высокий - **44,4%** Живут далеко от клуба, ходят в клубе без друзей, короткие абонементы, посещают клуб редко.

**Кластер 3**

**Самый высокий отток - 51,5%**. Живут рядом с клубом (судя по всему все из этого кластера), редко пользуются партнёрскими программами и "дружескими" акциями, самый низкий средний срок абонемента по всем кластерам (скорее всего купили абонемент на 1 месяц, чтобы попробовать, и не стали продлевать) и самый низкий срок "жизни" в клубе.

**Кластер 4**

Отток низкий - **6,8%**. Живут рядом с клубом, абонементы короткие, но ходят чаще, чем все остальные. Самый высокий срок "жизни" в клубе и самый высокий средний возраст.
